# Overview

The findings of this research will be valuable for students, educators, and policymakers seeking to understand the effectiveness of STEM programs at UC campuses and to identify potential areas for improvement.

9. Formatted Transfer Admission DFs (\t method same as rest)  
10. Filtered Transfer Admission DFs to only include TotalSum (or Total Enrollees) like Freshman

```python
# Total up all students of graduation year 20XX as estimate
total_enrolled_20 = fresh_adm_summary.loc[:, ('Campus', 'Enrollees 16')]
total_enrolled_20['tEnrollees 18'] = tran_adm_summary['tEnrollees 18']
total_enrolled_20['EstimatedTotalStudents'] = total_enrolled_20.sum(axis=1, numeric_only=True)
total_enrolled_21 = fresh_adm_summary.loc[:, ('Campus', 'Enrollees 17')]
total_enrolled_21['tEnrollees 19'] = tran_adm_summary['tEnrollees 19']
total_enrolled_21['EstimatedTotalStudents'] = total_enrolled_21.sum(axis=1, numeric_only=True)
total_enrolled_22 = fresh_adm_summary.loc[:, ('Campus', 'Enrollees 18')]
total_enrolled_22['tEnrollees 19'] = tran_adm_summary['tEnrollees 20']
total_enrolled_22['EstimatedTotalStudents'] = total_enrolled_22.sum(axis=1, numeric_only=True)

# Append awarded degrees
total_enrolled_20['STEM Degrees'] = award_data['Stem Degrees Awarded 20']
total_enrolled_20['Total Degrees'] = award_data['Total Degrees Awarded 20']
total_enrolled_21['STEM Degrees'] = award_data['Stem Degrees Awarded 21']
total_enrolled_21['Total Degrees'] = award_data['Total Degrees Awarded 21']
total_enrolled_22['STEM Degrees'] = award_data['Stem Degrees Awarded 22']
total_enrolled_22['Total Degrees'] = award_data['Total Degrees Awarded 22']

# Calculate Proportional STEM:Total Degree ratio
total_enrolled_20['STEMDegrees/TotalDegrees'] = total_enrolled_20['STEM Degrees']/total_enrolled_20['Total Degrees']
total_enrolled_21['STEMDegrees/TotalDegrees'] = total_enrolled_21['STEM Degrees']/total_enrolled_21['Total Degrees']
total_enrolled_22['STEMDegrees/TotalDegrees'] = total_enrolled_22['STEM Degrees']/total_enrolled_22['Total Degrees']

# Calculate Proportional Estimated Enrollment:Total Degree ratio
total_enrolled_20['EnrollToAward'] = total_enrolled_20['EstimatedTotalStudents']/total_enrolled_20['Total Degrees']
total_enrolled_21['EnrollToAward'] = total_enrolled_21['EstimatedTotalStudents']/total_enrolled_21['Total Degrees']
total_enrolled_22['EnrollToAward'] = total_enrolled_22['EstimatedTotalStudents']/total_enrolled_22['Total Degrees']

# Estimatedimated STEM grad counts
total_enrolled_20['EstimatedSTEMStudents'] = total_enrolled_20['EstimatedTotalStudents']*total_enrolled_20['EnrollToAward']*total_enrolled_20['STEMDegrees/TotalDegrees']
total_enrolled_21['EstimatedSTEMStudents'] = total_enrolled_21['EstimatedTotalStudents']*total_enrolled_21['EnrollToAward']*total_enrolled_21['STEMDegrees/TotalDegrees']
total_enrolled_22['EstimatedSTEMStudents'] = total_enrolled_22['EstimatedTotalStudents']*total_enrolled_22['EnrollToAward']*total_enrolled_22['STEMDegrees/TotalDegrees']

# Cast Int
total_enrolled_20['EstimatedSTEMStudents'] = total_enrolled_20['EstimatedSTEMStudents'].astype(int)
total_enrolled_21['EstimatedSTEMStudents'] = total_enrolled_21['EstimatedSTEMStudents'].astype(int)
total_enrolled_22['EstimatedSTEMStudents'] = total_enrolled_22['EstimatedSTEMStudents'].astype(int)

# Create yearly table
total_enrolled_by_year = total_enrolled_20.loc[:, ('Campus', 'EstimatedSTEMStudents', 'EstimatedTotalStudents', 'STEM Degrees', 'Total Degrees', 'STEMDegrees/TotalDegrees')]
total_enrolled_by_year = pd.concat([total_enrolled_by_year, total_enrolled_21.loc[:, ('Campus', 'EstimatedSTEMStudents', 'EstimatedTotalStudents', 'STEM Degrees', 'Total Degrees', 'STEMDegrees/TotalDegrees')]])
total_enrolled_by_year = pd.concat([total_enrolled_by_year, total_enrolled_22.loc[:, ('Campus', 'EstimatedSTEMStudents', 'EstimatedTotalStudents', 'STEM Degrees', 'Total Degrees', 'STEMDegrees/TotalDegrees')]])
total_enrolled_by_year['Year'] = [2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2021, 2021, 2021, 2021, 2021, 2021, 2021, 2021, 2021, 2022, 2022, 2022, 2022, 2022, 2022, 2022, 2022, 2022]
total_enrolled_by_year
```

# Conclusion & Discussion

From our analysis, our hypothesis that either UCLA or UC Berkeley will have the highest graduation rates among undergraduate students pursuing a degree in STEM fields was wrong. Surprisingly, our group concluded that the UC with the highest graduation rate among the STEM field (defined as the highest ratio of STEM degrees awarded to undergraduates compared to total degrees awarded) was UC Merced, at least across the graduating classes of 2020, 2021, and 2022. 

We approached our research question through the lens of proportional analysis, focusing primarily on the quantitative aspects of the data we collected. However, due to this method of analysis, there are a variety of factors that we failed to consider. One of the most notable variables influencing the data is age of campus. As the newest UC campus established, UC Merced has by far the smallest student population and the least admissions and graduates of all the UCs. This is the strongest explanation as to how UC Merced comes out on top as the UC with the highest graduation rate. Other variables that we did not consider for our research include program reputation, college rankings, and admission rates, to name a few. As such, it is important to note that the data analyzed focuses specifically on graduation rates, which is but one aspect of an effective education, and does not fully encompass all the reasons why a prospective student may want to choose a school. 

In conclusion, our analysis provides insights into the enrollment trends, STEM degree distribution, and proportional relationships between enrollment, degree awards, and STEM fields across different campuses and graduation years. It highlights the varying focus on STEM education and potential challenges related to STEM student retention. The findings can help inform decisions and strategies related to resource allocation, curriculum development, and support services aimed at promoting STEM education and improving student success in these fields.
